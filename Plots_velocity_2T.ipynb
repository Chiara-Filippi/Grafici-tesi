## Simulazioni phabs\*(bapec + bapec) per *Resolve*, *Xtend* e *XMM*

Sono sotto riportati i grafici, per diversi conteggi, di $T_{fit}/T_{input}$, della metallicità, di $N_{fit}/N_{input}$ e di $v_{fit}/v_{input}$ per la componente calda di uno spettro **phabs*(bapec + bapec)**, fittando un **phabs*(bapec)** con $T_{hot}$ e velocità impostate a 2 keV e a 300 km/s. 

I grafici sono raggruppati in base al numero di conteggi e divisi in due colonne in funzione della velocità di input.

$v_{fit}/v_{input}$ è rappresentata in scala logaritmica, per visualizzare meglio l'ordine di grandezza del rapporto.
