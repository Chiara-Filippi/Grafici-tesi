## Risoluzione della velocità per *Resolve*, *Xtend* e *XMM*.

Si osserva che solo *Resolve* riesce a risolvere le velocità per un numero di conteggi maggiore o uguale a 10000.

$v_{fit}/v_{input}$ è rappresentato in scala logaritmica, in modo da visualizzare meglio l'ordine di grandezza del rapporto.
