## Risoluzione della velocità per *Resolve*, *Xtend* e *XMM*.

Sono sotto riportati i grafici, per diversi conteggi, del rapporto tra la velocità ricavata dal fit e quella in input in funzione della velocità di input, simulando uno spettro **phabs*(bapec)** con temperatura e velocità variabili (rispettivamente [5, 10] keV e [100, 300, 500] km/s) e fittando un **phabs*(bapec)** con temperatura e velocità impostate a 5 keV e 300 km/s.

Si osserva che solo *Resolve* riesce a risolvere le velocità per un numero di conteggi maggiore o uguale a 10000.

$v_{fit}/v_{input}$ è rappresentato in scala logaritmica, in modo da visualizzare meglio l'ordine di grandezza del rapporto.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

fig, axes = plt.subplots(2, 2, figsize=(20, 20))

with open('Resolve/Velocity resolution/velocity_resolve_1000.dat') as f:
        lines = f.readlines()
        velocity = [float(line.split()[0]) for line in lines]
        temp_hot = [float(line.split()[1]) for line in lines]
        velocity_resolve = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_resolve = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]

with open('Xtend/Velocity resolution/velocity_xtend_1000.dat') as f:
        lines = f.readlines()
        velocity_xtend = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_xtend = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]

with open('XMM/Velocity resolution/velocity_xmm_1000.dat') as f:
        lines = f.readlines()
        velocity_xmm = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_xmm = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]
        
for j in range(len(temp_hot)):
    if temp_hot[j] == 5:
        if velocity[j] == 100.:
            axes[0, 0].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="P", markersize=10, label="$T_{hot} = 5$ keV, Xtend")
            axes[0, 0].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='cornflowerblue', marker="D", markersize=10, label="$T_{hot} = 5$ keV, XMM")
            axes[0, 0].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="o", markersize=10, label="$T_{hot} = 5$ keV, Resolve")
        axes[0, 0].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="o", markersize=10)
        axes[0, 0].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="P", markersize=10)
        axes[0, 0].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='cornflowerblue', marker="D", markersize=10)
        
    if temp_hot[j] == 10:
        if velocity[j] == 100.:
            axes[0, 0].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10, label="$T_{hot} = 10$ keV, Xtend")
            axes[0, 0].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='slateblue', marker="H",  markersize=10, label="$T_{hot} = 10$ keV, XMM")
            axes[0, 0].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="s", markersize=10, label="$T_{hot} = 10$ keV, Resolve")
        axes[0, 0].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="s", markersize=10)
        axes[0, 0].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10)
        axes[0, 0].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='slateblue', marker="H",  markersize=10)
        
with open('Resolve/Velocity resolution/velocity_resolve_10000.dat') as f:
        lines = f.readlines()
        velocity_resolve = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_resolve = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]

with open('Xtend/Velocity resolution/velocity_xtend_10000.dat') as f:
        lines = f.readlines()
        velocity_xtend = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_xtend = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]
        
with open('XMM/Velocity resolution/velocity_xmm_10000.dat') as f:
        lines = f.readlines()
        velocity_xmm = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_xmm = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]

for j in range(len(temp_hot)):
    if temp_hot[j] == 5:
        if velocity[j] == 100.:
            axes[0, 1].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="P", markersize=10, label="$T_{hot} = 5$ keV, Xtend")
            axes[0, 1].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='cornflowerblue', marker="D", markersize=10, label="$T_{hot} = 5$ keV, XMM")
            axes[0, 1].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="o", markersize=10, label="$T_{hot} = 5$ keV, Resolve")
        axes[0, 1].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="P", markersize=10)
        axes[0, 1].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='cornflowerblue', marker="D", markersize=10)
        axes[0, 1].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="o", markersize=10)
        
    if temp_hot[j] == 10:
        if velocity[j] == 100.:
            axes[0, 1].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10, label="$T_{hot} = 10$ keV, Xtend")
            axes[0, 1].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='slateblue', marker="H",  markersize=10, label="$T_{hot} = 10$ keV, XMM")
            axes[0, 1].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="s", markersize=10, label="$T_{hot} = 10$ keV, Resolve")
        axes[0, 1].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="s", markersize=10)
        axes[0, 1].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10)
        axes[0, 1].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='slateblue', marker="H",  markersize=10)

with open('Resolve/Velocity resolution/velocity_resolve_30000.dat') as f:
        lines = f.readlines()
        velocity = [float(line.split()[0]) for line in lines]
        temp_hot = [float(line.split()[1]) for line in lines]
        velocity_resolve = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_resolve = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]

with open('Xtend/Velocity resolution/velocity_xtend_30000.dat') as f:
        lines = f.readlines()
        velocity_xtend = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_xtend = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]

with open('XMM/Velocity resolution/velocity_xmm_30000.dat') as f:
        lines = f.readlines()
        velocity_xmm = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_xmm = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]
        
for j in range(len(temp_hot)):
    if temp_hot[j] == 5:
        if velocity[j] == 100.:
            axes[1, 0].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="P", markersize=10, label="$T_{hot} = 5$ keV, Xtend")
            axes[1, 0].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='cornflowerblue', marker="D", markersize=10, label="$T_{hot} = 5$ keV, XMM")
            axes[1, 0].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="o", markersize=10, label="$T_{hot} = 5$ keV, Resolve")
        axes[1, 0].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="o", markersize=10)
        axes[1, 0].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="P", markersize=10)
        axes[1, 0].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='cornflowerblue', marker="D", markersize=10)
        
    if temp_hot[j] == 10:
        if velocity[j] == 100.:
            axes[1, 0].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10, label="$T_{hot} = 10$ keV, Xtend")
            axes[1, 0].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='slateblue', marker="H",  markersize=10, label="$T_{hot} = 10$ keV, XMM")
            axes[1, 0].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="s", markersize=10, label="$T_{hot} = 10$ keV, Resolve")
        axes[1, 0].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="s", markersize=10)
        axes[1, 0].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10)
        axes[1, 0].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='slateblue', marker="H",  markersize=10)
        
with open('Resolve/Velocity resolution/velocity_resolve_100000.dat') as f:
        lines = f.readlines()
        velocity_resolve = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_resolve = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]

with open('Xtend/Velocity resolution/velocity_xtend_100000.dat') as f:
        lines = f.readlines()
        velocity_xtend = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_xtend = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]
        
with open('XMM/Velocity resolution/velocity_xmm_100000.dat') as f:
        lines = f.readlines()
        velocity_xmm = [float(line.split()[4])/float(line.split()[3]) for line in lines]
        error_xmm = [( float(line.split()[6]) - float(line.split()[5]) )/( 2*float(line.split()[3]) ) for line in lines]

for j in range(len(temp_hot)):
    if temp_hot[j] == 5:
        if velocity[j] == 100.:
            axes[1, 1].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="P", markersize=10, label="$T_{hot} = 5$ keV, Xtend")
            axes[1, 1].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='cornflowerblue', marker="D", markersize=10, label="$T_{hot} = 5$ keV, XMM")
            axes[1, 1].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="o", markersize=10, label="$T_{hot} = 5$ keV, Resolve")
        axes[1, 1].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="P", markersize=10)
        axes[1, 1].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='cornflowerblue', marker="D", markersize=10)
        axes[1, 1].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="o", markersize=10)
        
    if temp_hot[j] == 10:
        if velocity[j] == 100.:
            axes[1, 1].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10, label="$T_{hot} = 10$ keV, Xtend")
            axes[1, 1].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='slateblue', marker="H",  markersize=10, label="$T_{hot} = 10$ keV, XMM")
            axes[1, 1].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="s", markersize=10, label="$T_{hot} = 10$ keV, Resolve")
        axes[1, 1].errorbar(velocity[j]+15, velocity_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="s", markersize=10)
        axes[1, 1].errorbar(velocity[j]-15, velocity_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10)
        axes[1, 1].errorbar(velocity[j], velocity_xmm[j], yerr=error_xmm[j], color='slateblue', marker="H",  markersize=10)
        
for k in [0, 1]:
    for l in [0, 1]:
        axes[k, l].set_xlabel('$V_{input}$', fontsize = 15)
        axes[k, l].legend()
        axes[k, l].grid(True)
        axes[k, l].set_ylabel('$V_{fit}/V_{input}$', fontsize = 15)
        axes[k, l].set_yscale('log')
        axes[k, l].axhline(y=1, color='black')

axes[0, 0].set_title('1000 conteggi', fontsize = 15)
axes[0, 1].set_title('10000 conteggi', fontsize = 15)
axes[1, 0].set_title('30000 conteggi', fontsize = 15)
axes[1, 1].set_title('100000 conteggi', fontsize = 15)

    
# Show the plot
plt.show()
```
