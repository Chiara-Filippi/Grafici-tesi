## Confronto tra _Resolve_ e _Xtend_ per $T_{hot} = 5 $ keV e $T_{hot} = 10$ keV

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

with open('Resolve/temp_resolve.dat') as f:
        lines = f.readlines()
        temp_cold = [float(line.split()[0]) for line in lines]
        temp_hot = [float(line.split()[1]) for line in lines]
        norm = [float(line.split()[2])/float(line.split()[3]) for line in lines]
        ratio_resolve = [float(line.split()[5])/float(line.split()[4]) for line in lines]
        error_resolve = [( float(line.split()[7]) - float(line.split()[6]) )/( 2*float(line.split()[4]) ) for line in lines]

with open('Xtend/temp_xtend.dat') as f:
        lines = f.readlines()
        ratio_xtend = [float(line.split()[5])/float(line.split()[4]) for line in lines]
        error_xtend = [( float(line.split()[7]) - float(line.split()[6]) )/( 2*float(line.split()[4]) ) for line in lines]

i = 0;
fig, axes = plt.subplots(3, 2, figsize=(20, 30))

for j in range(len(temp_hot)):
    if temp_hot[j] == 5:
        i = 0
    if temp_hot[j] == 10:
        i = 1
    if temp_cold[j] == 0.5:
        if round(norm[j],2) == 0.05:
            axes[0, i].errorbar(norm[j]+0.01, ratio_resolve[j], yerr=error_resolve[j], color='blueviolet', marker="o", markersize=10, label="$T_{cold}=0.5$ keV, Resolve")
            axes[0, i].errorbar(norm[j]-0.01, ratio_xtend[j], yerr=error_xtend[j], color='mediumturquoise', marker="P", markersize=10, label="$T_{cold}=0.5$ keV, Xtend")
        else:
            axes[0, i].errorbar(norm[j]+0.01, ratio_resolve[j], yerr=error_resolve[j], color='blueviolet', marker="o", markersize=10)
            axes[0, i].errorbar(norm[j]-0.01, ratio_xtend[j], yerr=error_xtend[j], color='mediumturquoise', marker="P", markersize=10)
    if temp_cold[j] == 1.:
        if round(norm[j],2) == 0.05:
            axes[0, i].errorbar(norm[j]+0.01, ratio_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="s",  markersize=10, label="$T_{cold}=1$ keV, Resolve")
            axes[0, i].errorbar(norm[j]-0.01, ratio_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10, label="$T_{cold}=1$ keV, Xtend")
        else:
            axes[0, i].errorbar(norm[j]+0.01, ratio_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="s", markersize=10)
            axes[0, i].errorbar(norm[j]-0.01, ratio_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10)
    if temp_cold[j] == 2.:
        if round(norm[j],2) == 0.05:
            axes[0, i].errorbar(norm[j]+0.01, ratio_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="D", markersize=10, label="$T_{cold}=2$ keV, Resolve")
            axes[0, i].errorbar(norm[j]-0.01, ratio_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="H", markersize=10, label="$T_{cold}=2$ keV, Xtend")
        else:
            axes[0, i].errorbar(norm[j]+0.01, ratio_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="D", markersize=10)
            axes[0, i].errorbar(norm[j]-0.01, ratio_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="H", markersize=10)

with open('Resolve/abun_resolve.dat') as f:
        lines = f.readlines()
        abun_resolve = [float(line.split()[5]) for line in lines]
        error_resolve = [( float(line.split()[7]) - float(line.split()[6]) )/2 for line in lines]


with open('Xtend/abun_xtend.dat') as f:
        lines = f.readlines()
        abun_xtend = [float(line.split()[5]) for line in lines]
        error_xtend = [( float(line.split()[7]) - float(line.split()[6]) )/2 for line in lines]
i = 0;

for j in range(len(temp_hot)):
    if temp_hot[j] == 5:
        i = 0
    if temp_hot[j] == 10:
        i = 1
    if temp_cold[j] == 0.5:
        if round(norm[j],2) == 0.05:
            axes[1, i].errorbar(norm[j]+0.01, abun_resolve[j], yerr=error_resolve[j], color='blueviolet', marker="o", markersize=10, label="$T_{cold}=0.5$ keV, Resolve")
            axes[1, i].errorbar(norm[j]-0.01, abun_xtend[j], yerr=error_xtend[j], color='mediumturquoise', marker="P", markersize=10, label="$T_{cold}=0.5$ keV, Xtend")
        else:
            axes[1, i].errorbar(norm[j]+0.01, abun_resolve[j], yerr=error_resolve[j], color='blueviolet', marker="o", markersize=10)
            axes[1, i].errorbar(norm[j]-0.01, abun_xtend[j], yerr=error_xtend[j], color='mediumturquoise', marker="P", markersize=10)
    if temp_cold[j] == 1.:
        if round(norm[j],2) == 0.05:
            axes[1, i].errorbar(norm[j]+0.01, abun_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="s",  markersize=10, label="$T_{cold}=1$ keV, Resolve")
            axes[1, i].errorbar(norm[j]-0.01, abun_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10, label="$T_{cold}=1$ keV, Xtend")
        else:
            axes[1, i].errorbar(norm[j]+0.01, abun_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="s", markersize=10)
            axes[1, i].errorbar(norm[j]-0.01, abun_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10)
    if temp_cold[j] == 2.:
        if round(norm[j],2) == 0.05:
            axes[1, i].errorbar(norm[j]+0.01, abun_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="D", markersize=10, label="$T_{cold}=2$ keV, Resolve")
            axes[1, i].errorbar(norm[j]-0.01, abun_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="H", markersize=10, label="$T_{cold}=2$ keV, Xtend")
        else:
            axes[1, i].errorbar(norm[j]+0.01, abun_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="D", markersize=10)
            axes[1, i].errorbar(norm[j]-0.01, abun_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="H", markersize=10)


with open('Resolve/norm_resolve.dat') as f:
    lines = f.readlines()
    norm_resolve = [float(line.split()[4]) for line in lines]
    error_resolve = [( float(line.split()[6]) - float(line.split()[5]) )/2 for line in lines]


with open('Xtend/norm_xtend.dat') as f:
        lines = f.readlines()
        norm_xtend = [float(line.split()[4]) for line in lines]
        error_xtend = [( float(line.split()[6]) - float(line.split()[5]) )/2 for line in lines]
i = 0;

for j in range(len(temp_hot)):
    if temp_hot[j] == 5:
        i = 0
    if temp_hot[j] == 10:
        i = 1
    if temp_cold[j] == 0.5:
        if round(norm[j],2) == 0.05:
            axes[2, i].errorbar(norm[j]+0.01, norm_resolve[j], yerr=error_resolve[j], color='blueviolet', marker="o", markersize=10, label="$T_{cold}=0.5$ keV, Resolve")
            axes[2, i].errorbar(norm[j]-0.01, norm_xtend[j], yerr=error_xtend[j], color='mediumturquoise', marker="P", markersize=10, label="$T_{cold}=0.5$ keV, Xtend")
        else:
            axes[2, i].errorbar(norm[j]+0.01, norm_resolve[j], yerr=error_resolve[j], color='blueviolet', marker="o", markersize=10)
            axes[2, i].errorbar(norm[j]-0.01, norm_xtend[j], yerr=error_xtend[j], color='mediumturquoise', marker="P", markersize=10)
    if temp_cold[j] == 1.:
        if round(norm[j],2) == 0.05:
            axes[2, i].errorbar(norm[j]+0.01, norm_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="s",  markersize=10, label="$T_{cold}=1$ keV, Resolve")
            axes[2, i].errorbar(norm[j]-0.01, norm_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10, label="$T_{cold}=1$ keV, Xtend")
        else:
            axes[2, i].errorbar(norm[j]+0.01, norm_resolve[j], yerr=error_resolve[j], color='mediumorchid', marker="s", markersize=10)
            axes[2, i].errorbar(norm[j]-0.01, norm_xtend[j], yerr=error_xtend[j], color='mediumseagreen', marker="X",  markersize=10)
    if temp_cold[j] == 2.:
        if round(norm[j],2) == 0.05:
            axes[2, i].errorbar(norm[j]+0.01, norm_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="D", markersize=10, label="$T_{cold}=2$ keV, Resolve")
            axes[2, i].errorbar(norm[j]-0.01, norm_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="H", markersize=10, label="$T_{cold}=2$ keV, Xtend")
        else:
            axes[2, i].errorbar(norm[j]+0.01, norm_resolve[j], yerr=error_resolve[j], color='mediumvioletred', marker="D", markersize=10)
            axes[2, i].errorbar(norm[j]-0.01, norm_xtend[j], yerr=error_xtend[j], color='yellowgreen', marker="H", markersize=10)

for k in [0, 1]:
    for l in [0, 1, 2]:
        axes[l, k].set_xlabel('$N_{cold}/N_{hot}$', fontsize = 15)
        axes[l, k].legend()
        axes[l, k].grid(True)
        axes[l, 0].set_title('$T_{hot} = 5$ keV', fontsize = 15)
        axes[l, 1].set_title('$T_{hot} = 10$ keV', fontsize = 15)
    axes[0, k].set_ylabel('$T_{fit}/T_{input}$', fontsize = 15)
    axes[1, k].set_ylabel('Metallicità', fontsize = 15)
    axes[2, k].set_ylabel('$N_{hot}$', fontsize = 15)
```
