## Simulazioni phabs\*(apec + apec) per *Resolve*, *Xtend* e *XMM*.

Sono sotto riportati i grafici, per diversi conteggi, di $T_{fit}/T_{input}$, della metallicità, di $N_{fit}/N_{input}$ per la componente calda di uno spettro **phabs*(apec + apec)**, fittando un **phabs*(apec)** con $T_{hot}$ e velocità impostate a 2 keV. 

Si nota una forte anti-correlazione tra $T_{fit}/T_{input}$ e $N_{fit}/N_{input}$. Per *Resolve* si osserva che la componente fredda che più abbassa $T_{fit}$ è quella a $T_{cold}$=1 keV.
