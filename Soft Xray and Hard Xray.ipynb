## Simulazioni phabs\*(apec + apec) per *Resolve*, *Xtend* e *XMM*.

Sono sotto riportati i grafici, per diversi conteggi, di $T_{fit}/T_{input}$, della metallicità, di $N_{fit}/N_{input}$ per la componente fredda di uno spettro **phabs*(apec + apec)**, fittando un **phabs*(apec)** nel range 0.5-2 keV con temperatura del modello fissata a 2 keV (**banda soft**).

Si nota che la temperatura della componente fredda viene recuperata solo quando la sua normalizzazione è alta (50% di quella calda) e che vi è una correlazione diretta tra $T_{fit}/T_{input}$ e $N_{fit}/N_{input}$: la normalizzazione viene alzata quando il fit non riesce ad avvicinarsi alla temperatura corretta.
